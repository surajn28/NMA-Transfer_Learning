```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#filenames = glob.glob("../input/inceptionresnetv1-vgg-face2-to-kdef/*.csv")
```

```python
results = pd.DataFrame()

current_results = pd.read_csv("../input/transfer-learning-comparative/InceptionResnetV1_untrained_to_kdef__lr_0.01__bs_32(1).csv", sep = ",")
current_test_accuracy = current_results['test acc'].values
results["control - kdef from scratch"] = current_test_accuracy

current_results = pd.read_csv("../input/transfer-learning-comparative/InceptionResnetV1_vgg_face2_to_kdef__lr_0.0005__bs_32.csv", sep = ",")
current_test_accuracy = current_results['test acc'].values
results["transfer - only last layer"] = current_test_accuracy

current_results = pd.read_csv("../input/transfer-learning-comparative/InceptionResnetV1_vgg_face2_to_kdef__lr_0.01__bs_32_full-retrain_100ep.csv", sep = ",")
current_test_accuracy = current_results['test acc'].values
results["transfer - all layers"] = current_test_accuracy
```

```python
results
```

```python
figureName = "transfer_learning_comparision"

plt.figure(figsize=(10,5))
plt.plot(range(0, 100), results, alpha = 0.75, label=results.columns)
plt.xlabel('Number of epochs')
plt.ylabel('Accuracy')
plt.title(f'Test Accuracy Curve')
plt.legend(title = "Models")
plt.savefig(figureName)
plt.show()
```

```python
#results.max(axis=0)
stats = pd.DataFrame()
stats["lr"] = results.columns
stats["lr"] = stats["lr"].astype("float")

#for col in results.columns:
#    print(results.nlargest(3, col)[col].mean())
    
stats["mean"] = results.mean().tolist()
stats["max"] = results.max().tolist()
stats["mean_max_3"] = [results.nlargest(3, col)[col].mean() for col in results.columns]
```

```python
stats
```

```python
stats = stats.sort_values("lr")
print(stats)
```

```python
plt.figure(figsize=(10,5))
plt.plot(stats["lr"], stats["mean"], alpha = 0.5, label="mean")
plt.plot(stats["lr"], stats["max"], alpha = 0.5, label="max")
plt.plot(stats["lr"], stats["mean_max_3"], alpha = 0.5, label="mean_max_3")
plt.xscale('log')
plt.xlabel('lr')
plt.ylabel('Accuracy')
#plt.title(f'Train/Test Accuracy curve for {max_epochs} epochs')
#plt.savefig(figureName)
plt.legend()
plt.show()
```
